```python
# Question 1 Answer

class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

        
def get_greater_node(node1, node2):
    if node1.value >= node2.value:
        return node1
    else:
        return node2

def merge_linked_lists(head1, head2):
    new_head = None
    new_tail = None

    while head1 is not None and head2 is not None:
        greater_node = get_greater_node(head1, head2)

        if new_head is None:
            new_head = greater_node
            new_tail = greater_node
        else:
            new_tail.next = greater_node
            new_tail = greater_node

        head1 = head1.next
        head2 = head2.next

    # If any of the lists still has remaining nodes, append them to the new list
    if head1 is not None:
        if new_head is None:
            new_head = head1
        else:
            new_tail.next = head1

    if head2 is not None:
        if new_head is None:
            new_head = head2
        else:
            new_tail.next = head2

    return new_head

# Create the first linked list: 1 -> 3 -> 5 -> 7 -> None
head1 = Node(1)
head1.next = Node(3)
head1.next.next = Node(5)
head1.next.next.next = Node(7)

# Create the second linked list: 2 -> 4 -> 6 -> 8 -> None
head2 = Node(2)
head2.next = Node(4)
head2.next.next = Node(6)
head2.next.next.next = Node(8)

# Merge the two linked lists based on the condition
new_head = merge_linked_lists(head1, head2)

# Print the resulting linked list
current_node = new_head
while current_node is not None:
    print(current_node.value, end=" -> ")
    current_node = current_node.next
print("None")
```

```python
# Question 2 Answer


def remove_duplicates(head):
    if head is None:
        return head

    current = head

    while current.next is not None:
        if current.value == current.next.value:
            current.next = current.next.next
        else:
            current = current.next

    return head
# Create the linked list: 11 -> 11 -> 11 -> 21 -> 43 -> 43 -> 60 -> None
head = Node(11)
head.next = Node(11)
head.next.next = Node(11)
head.next.next.next = Node(21)
head.next.next.next.next = Node(43)
head.next.next.next.next.next = Node(43)
head.next.next.next.next.next.next = Node(60)

# Remove duplicate nodes
new_head = remove_duplicates(head)

# Print the resulting linked list
current_node = new_head
while current_node is not None:
    print(current_node.value, end=" -> ")
    current_node = current_node.next
print("None")
```

```python
# Question 5 Answer

def delete_last_occurrence(head, key):
    if head is None:
        return head

    # Find the last occurrence of the key
    last_occurrence = None
    prev = None
    current = head

    while current is not None:
        if current.value == key:
            last_occurrence = current
        prev = current
        current = current.next

    # If the last occurrence is found, delete it
    if last_occurrence is not None:
        if prev is None:
            head = last_occurrence.next
        else:
            prev.next = last_occurrence.next

    return head
# Create the linked list: 1 -> 2 -> 3 -> 2 -> 4 -> 5 -> 2 -> None
head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(2)
head.next.next.next.next = Node(4)
head.next.next.next.next.next = Node(5)
head.next.next.next.next.next.next = Node(2)

# Delete the last occurrence of key 2
new_head = delete_last_occurrence(head, 2)

# Print the resulting linked list
current_node = new_head
while current_node is not None:
    print(current_node.value, end=" -> ")
    current_node = current_node.next
print("None")
```

```python
# Question 6 Answer

def merge_sorted_lists(head1, head2):
    if head1 is None:
        return head2
    if head2 is None:
        return head1

    # Determine the head of the merged list
    if head1.value <= head2.value:
        merged_head = head1
        head1 = head1.next
    else:
        merged_head = head2
        head2 = head2.next

    current = merged_head

    # Merge the two lists
    while head1 is not None and head2 is not None:
        if head1.value <= head2.value:
            current.next = head1
            head1 = head1.next
        else:
            current.next = head2
            head2 = head2.next
        current = current.next

    # Append any remaining nodes from either list
    if head1 is not None:
        current.next = head1
    if head2 is not None:
        current.next = head2

    return merged_head
# Create the first sorted linked list: 1 -> 3 -> 5 -> 7 -> None
head1 = Node(1)
head1.next = Node(3)
head1.next.next = Node(5)
head1.next.next.next = Node(7)

# Create the second sorted linked list: 2 -> 4 -> 6 -> 8 -> None
head2 = Node(2)
head2.next = Node(4)
head2.next.next = Node(6)
head2.next.next.next = Node(8)

# Merge the two sorted linked lists
merged_head = merge_sorted_lists(head1, head2)

# Print the resulting merged linked list
current_node = merged_head
while current_node is not None:
    print(current_node.value, end=" -> ")
    current_node = current_node.next
print("None")
```

```python
# Question 7 Answer

class Node:
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None

def reverse_doubly_linked_list(head):
    if head is None or head.next is None:
        return head

    current = head
    while current is not None:
        # Swap the previous and next pointers of the current node
        current.prev, current.next = current.next, current.prev

        # Move to the next node
        head = current
        current = current.prev

    return head
# Create the doubly linked list: 1 <-> 2 <-> 3 <-> 4 <-> 5
head = Node(1)
head.next = Node(2)
head.next.prev = head
head.next.next = Node(3)
head.next.next.prev = head.next
head.next.next.next = Node(4)
head.next.next.next.prev = head.next.next
head.next.next.next.next = Node(5)
head.next.next.next.next.prev = head.next.next.next

# Reverse the doubly linked list
reversed_head = reverse_doubly_linked_list(head)

# Print the resulting reversed doubly linked list
current_node = reversed_head
while current_node is not None:
    print(current_node.value, end=" <-> ")
    current_node = current_node.next
print("None")
```

```python
# Question 8 Answer

def delete_node_at_position(head, position):
    if head is None:
        return head

    # If position is 0, delete the head node
    if position == 0:
        head = head.next
        if head is not None:
            head.prev = None
        return head

    current = head
    count = 0

    # Traverse to the node at the given position
    while current is not None and count < position:
        current = current.next
        count += 1

    # If current is None, position is out of bounds
    if current is None:
        return head

    # Adjust the pointers to delete the node
    if current.prev is not None:
        current.prev.next = current.next
    if current.next is not None:
        current.next.prev = current.prev

    return head

# Create the doubly linked list: 1 <-> 2 <-> 3 <-> 4 <-> 5
head = Node(1)
head.next = Node(2)
head.next.prev = head
head.next.next = Node(3)
head.next.next.prev = head.next
head.next.next.next = Node(4)
head.next.next.next.prev = head.next.next
head.next.next.next.next = Node(5)
head.next.next.next.next.prev = head.next.next.next

# Delete the node at position 2
modified_head = delete_node_at_position(head, 2)

# Print the resulting modified doubly linked list
current_node = modified_head
while current_node is not None:
    print(current_node.value, end=" <-> ")
    current_node = current_node.next
print("None")
```
